```python
#--------------------------------------------------------------------------------------
# .                    PURCHASE ANALYSIS  by GENDER
# .   Two different methods are uised:
# .                    1. The long method! is to create 3 separate data frames for each group
#                       and derive the necessary stats for each data frame + report
#                      2. The second method is create a group data and then a summary table
#
#       First the long method!
#--------------------------------------------------------------------------------------
# .  Create 3 different Data Frames each of which will be used separately in the program
#---------------------------------------------------------------------------------------
maleDF = hp.loc[hp["Gender"]=="Male"]
maleDF.count()
femaleDF = hp.loc[hp["Gender"]=="Female"]
femaleDF.count()
otherDF = hp.loc[hp["Gender"]=="Other / Non-Disclosed"]
otherDF["Gender"].count()
#hp.count()
#
#
```

```python
# Reference specific column within a DataFrame
# hp[["Price","Gender"]].head(10)
```

```python
#----------------------------------------------------------------------------
#           Calulate the % distribution of Males, Females & Others
#           Check that sum of these weights add up to 1
#----------------------------------------------------------------------------
maleCount=maleDF["Gender"].count()
femaleCount=femaleDF["Gender"].count()
otherCount=otherDF["Gender"].count()
total=hp["Gender"].count()
#----------------------------------------------------------------------------
percMaleCount = maleDF["Gender"].count()/hp["Gender"].count()
percFemaleCount = femaleDF["Gender"].count()/hp["Gender"].count()
percOtherCount = otherDF["Gender"].count()/hp["Gender"].count()
percMaleCount,percFemaleCount,percOtherCount
sumOfWeights = percMaleCount + percFemaleCount + percOtherCount
sumOfWeights

# Calculations can also be performed on Series and added into DataFrames as new columns
# In this case the percentage of Gender is used in the aggregation by Gender. 
genderCount = 1
hp["Gender Count"] = genderCount
hp["percGender"] = 1/total
```

```python
#-----------------------------------------------------------------------------------------------------
#                         Report the above results
#-----------------------------------------------------------------------------------------------------
# Initialise the results to a list (Data2) of dictionaries consisting of [labels:Values] for reporting 
data2 = [{'Number of males': maleCount,'%Male': percMaleCount,'Number of females': femaleCount, 
          '%Female': percFemaleCount ,'Other': otherCount, 
          '%Other': percOtherCount}]  
#-----------------------------------------------------------------------------------------------------
#                        Creates a New Data Frame with above data + + +  Formats the Result
#-----------------------------------------------------------------------------------------------------
newdf = pd.DataFrame(data2) 
# formats the selected data in the data frame for the report self
newdf["%Male"] = newdf['%Male'].astype(float).map("{:,.2%}".format)
newdf["%Female"] = newdf['%Female'].astype(float).map("{:,.2%}".format)
newdf["%Other"] = newdf['%Other'].astype(float).map("{:,.2%}".format)
#-----------------------------------------------------------------------------------------------------
newdf
```
